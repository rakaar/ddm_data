```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ILD_arr = np.arange(-16, 16, 1)
rate_lambda = 0.1
Gamma_non_lin = np.tanh(rate_lambda * ILD_arr / 17.37)
Gamma_lin = rate_lambda * ILD_arr / 17.37
```

```python
plt.plot(ILD_arr, Gamma_lin, label='linear')
plt.plot(ILD_arr, Gamma_non_lin, label='non linear', ls='--', lw=3)
plt.legend()
```

```python
right_level_arr = np.arange(0, 70, 10)
r_by_r0_non_norm = 10**(rate_lambda*right_level_arr / 20)
plt.plot(right_level_arr, r_by_r0_non_norm)

left_level = 35
ABLs = (right_level_arr + left_level) / 2
ILDs = right_level_arr - left_level

numerator = 10**(rate_lambda * right_level_arr / 20)
l = 0.95 
denom = 2* (10 **(rate_lambda* l * ABLs/20)) * np.cosh(rate_lambda * ILDs * l / 17.37)
norm_evid = numerator / denom
plt.plot(right_level_arr,norm_evid)
plt.ylim(0,2)
```

```python
rate_lambda = 2 
right_level_arr = np.arange(0, 100, 10)
r_by_r0_non_norm = 10**(rate_lambda*right_level_arr / 20)
# plt.plot(right_level_arr, r_by_r0_non_norm, label='no norm')

left_level = 35
l = 0.95
print(f'lambda x l = {rate_lambda * l}')
denom = (10**(rate_lambda*right_level_arr*l / 20)) + (10**(rate_lambda*left_level*l / 20))

norm_evid = r_by_r0_non_norm / denom
plt.plot(right_level_arr, norm_evid, label='norm')
plt.xlabel('right pressure')
plt.ylabel('r+/r0')

plt.axvline(left_level, alpha=0.2, color='k', ls='--', label='left channel')
plt.axhline(0.5, alpha=0.2, color='k', ls='--')
plt.legend();
```
